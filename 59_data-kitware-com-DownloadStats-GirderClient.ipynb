```python
API_URL = "https://data.kitware.com/api/v1"
import girder_client
gc = girder_client.GirderClient(apiUrl=API_URL)
```

```python
topLevelFolder = "5898b7ef8d777f07219fcb14" # NIH Shape Analysis Toolbox / SlicerSALT-Public-Packages
```

```python
def displayFileDownloadStats(item, level=0):
    if isinstance(item, str):
        item = gc.getItem(item)
    itemId = item["_id"]
    for file in gc.listFile(itemId):
        downloadStatistics = file.get("downloadStatistics", {"completed": 0})
        print(f"{level * ' '}* {file['name']}: {downloadStatistics['completed']}")

def displayFolderDownloadStats(folder, level=0):
    if isinstance(folder, str):
        folder = gc.getFolder(folder)
    print(f"{level * ' '}{folder['name']}:")
    folderId = folder["_id"]
    for item in gc.listItem(folderId):
        displayFileDownloadStats(item, level)
    for folder in gc.listFolder(folderId):
        displayFolderDownloadStats(folder, level + 2)
```

```python
displayFolderDownloadStats(topLevelFolder)
```
